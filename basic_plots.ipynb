```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load Iris data
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# Standardize the data
X_std = StandardScaler().fit_transform(X)

# Create subplots: 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 1 Box Plot (for one feature, e.g., petal length)
axes[0].boxplot(X_std[:, 2], vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
axes[0].set_title('Box Plot - Petal Length (standardized)')
axes[0].set_ylabel('Standardized Value')
axes[0].set_xticks([1])
axes[0].set_xticklabels(['Petal Length'])

# 2 Scatter Plot (e.g., Petal Length vs Petal Width)
axes[1].scatter(X_std[:, 2], X_std[:, 3], c=y, cmap='viridis', edgecolor='k', s=60)
axes[1].set_title('Scatter Plot - Petal Length vs Petal Width')
axes[1].set_xlabel('Petal Length (std)')
axes[1].set_ylabel('Petal Width (std)')

# Improve layout
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset and convert to DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Plot: Violin plot of petal length by species
plt.figure(figsize=(8, 5))
sns.violinplot(x='species', y='petal length (cm)', data=df, palette='Pastel1')
plt.title('Distribution of Petal Length by Species (Violin Plot)')
plt.ylabel('Petal Length (cm)')
plt.xlabel('Species')
plt.tight_layout()
plt.show()
```
